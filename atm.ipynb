```python
def add_new_user(conn):
   c = conn.cursor()

   # Get user details
   customer_name = input("Enter customer name: ")
   customer_address = input("Enter customer address: ")
   customer_contact_no = input("Enter customer contact number: ")

   # Insert new user into Customer table
   c.execute("INSERT INTO Customer (customer_name, customer_address, customer_contact_no) VALUES (?, ?, ?)",
             (customer_name, customer_address, customer_contact_no))

   conn.commit()

   print("New user added successfully.")


def display_all_customers(conn):
  c = conn.cursor()

  # Execute the SELECT statement
  c.execute("SELECT * FROM Customer")

  # Fetch all the records
  records = c.fetchall()

  # Print all the records
  for record in records:
      print(f"Customer ID: {record[0]}, Name: {record[1]}, Address: {record[2]}, Contact No: {record[3]}")


def search_customer(conn):
  c = conn.cursor()

  # Get customer ID
  customer_id = input("Enter customer ID: ")

  # Execute the SELECT statement
  c.execute("SELECT * FROM Customer WHERE customer_id=?", (customer_id,))

  # Fetch the record
  record = c.fetchone()

  # Print the record
  if record:
      print(f"Customer ID: {record[0]}, Name: {record[1]}, Address: {record[2]}, Contact No: {record[3]}")
  else:
      print("No customer found with the given ID.")

def open_new_account(conn):
  c = conn.cursor()

  # Get account details
  customer_id = input("Enter customer ID: ")
  account_number = input("Enter account number: ")
  pin = input("Enter 4-digit PIN: ")
  account_type = input("Enter account type (saving/current): ")
  amount = float(input("Enter initial amount: "))

  # Insert new account into Account table
  c.execute("INSERT INTO Account (customer_id, account_number, pin, account_type, amount) VALUES (?, ?, ?, ?, ?)",
            (customer_id, account_number, pin, account_type, amount))

  conn.commit()

  print("New account opened successfully.")

def display_all_accounts(conn):
 c = conn.cursor()

 # Execute the SELECT statement
 c.execute("SELECT * FROM Account")

 # Fetch all the records
 records = c.fetchall()

 # Print all the records
 for record in records:
     print(f"Account ID: {record[0]}, Customer ID: {record[1]}, Account Number: {record[2]}, PIN: {record[3]}, Account Type: {record[4]}, Amount: {record[5]}")

def search_account(conn):
 c = conn.cursor()

 # Get account number
 account_number = input("Enter account number: ")

 # Execute the SELECT statement
 c.execute("SELECT * FROM Account WHERE account_number=?", (account_number,))

 # Fetch the record
 record = c.fetchone()

 # Print the record
 if record:
     print(f"Account ID: {record[0]}, Customer ID: {record[1]}, Account Number: {record[2]}, PIN: {record[3]}, Account Type: {record[4]}, Amount: {record[5]}")
 else:
     print("No account found with the given number.")

def withdraw_amount(conn):
 c = conn.cursor()

 # Get account number and withdrawal amount
 account_number = input("Enter account number: ")
 withdrawal_amount = float(input("Enter amount to withdraw: "))

 # Execute the UPDATE statement
 c.execute("SELECT amount FROM Account WHERE account_number=?", (account_number,))
 record = c.fetchone()

 if record:
     current_amount = record[0]
     if current_amount >= withdrawal_amount:
         c.execute("UPDATE Account SET amount=amount-? WHERE account_number=?", (withdrawal_amount, account_number))
         conn.commit()
         print("Withdrawal successful.")
     else:
         print("Insufficient balance.")
 else:
     print("No account found with the given number.")

def help_function():
 print("Welcome to the ATM Management System. Here are the options you can choose from:")
 print("1: Add a new user - Prompts you to enter the customer's name, address, and contact number. This information is stored in the Customer table in the database.")
 print("2: Display all customers - Retrieves and prints all the customers from the Customer table.")
 print("3: Search for a customer by ID - Prompts you to enter a customer ID. The system retrieves and prints the customer with the matching ID from the Customer table.")
 print("4: Open a new account - Prompts you to enter the customer ID, account number, 4-digit PIN, account type (saving or current), and initial amount. This information is stored in the Account table in the database.")
 print("5: Display all accounts - Retrieves and prints all the accounts from the Account table.")
 print("6: Search for an account by number - Prompts you to enter an account number. The system retrieves and prints the account with the matching number from the Account table.")
 print("7: Withdraw an amount from an account - Prompts you to enter an account number and the amount to withdraw. The system checks if the account has sufficient balance. If it does, the system decreases the amount in the Account table by the withdrawal amount.")
 print("8: Exit the system - Closes the database connection and ends the program.")
 print("0: Display this help message - Prints this help message that explains how to use the system.")

def main(conn):

  while True:
    print("\n!=========================*************=======================!")
    print("! PLEASE ENTER 0 FOR HELP")
    print("! PLEASE ENTER 1 FOR NEW USER")
    print("! PLEASE ENTER 2 TO DISPLAY ALL CUSTOMERS")
    print("! PLEASE ENTER 3 TO SEARCH A CUSTOMER")
    print("! PLEASE ENTER 4 TO OPEN NEW ACCOUNT")
    print("! PLEASE ENTER 5 TO DISPLAY ALL ACCOUNTS")
    print("! PLEASE ENTER 6 TO SEARCH AN ACCOUNT")
    print("! PLEASE ENTER 7 TO WITHDRAW AMOUNT")
    print("! PLEASE ENTER 8 TO EXIT")
    print("\n!============================*****END*****====================!")

    choice = input("Enter your choice: ")

    if choice == '1':
      add_new_user(conn)
    elif choice == '2':
      display_all_customers(conn)
    elif choice == '3':
      search_customer(conn)
    elif choice == '4':
      open_new_account(conn)
    elif choice == '5':
      display_all_accounts(conn)
    elif choice == '6':
      search_account(conn)
    elif choice == '7':
      withdraw_amount(conn)
    elif choice == '8':
      print("Thank you.")
      break
    elif choice == '0':
      help_function()
    else:
      print("Invalid choice")

if __name__ == "__main__":

  print("ATM MANAGEMENT SYSTEM")
  print("Created and maintained by: Nasya Verma, Sri Chaitanya Techno School")
  print("")
  print("")
  print("Enter database name:")

  db_name = input()
  conn = sqlite3.connect(db_name)
  c = conn.cursor()

  print("SQLite Database connection successful.")

  # Create Customer table
  c.execute('''CREATE TABLE IF NOT EXISTS Customer
  (customer_id INTEGER PRIMARY KEY,
  customer_name TEXT NOT NULL,
  customer_address TEXT,
  customer_contact_no TEXT NOT NULL)''')

  # Create Account table
  c.execute('''CREATE TABLE IF NOT EXISTS Account
  (account_id INTEGER PRIMARY KEY,
  customer_id INTEGER,
  account_number TEXT NOT NULL,
  pin TEXT CHECK(LENGTH(pin) = 4),
  account_type TEXT NOT NULL,
  amount REAL,
  FOREIGN KEY (customer_id) REFERENCES Customer(customer_id))''')

  conn.commit()

  main(conn)
```
